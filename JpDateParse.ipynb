```python
html = urllib.urlopen(target)
bsObj = BeautifulSoup(html.read())

print(bsObj.h1)
print(bsObj.title)
```

```python
rawStr = str(bsObj.title)

p = re.compile(r"\d+") #把所有的數字都抓出來

match = p.findall(rawStr) #這個很好用，回傳 list，在這裡 match[0] = year, match[1] = month, match[2] = date
"""
 if match:
    # 得到匹配結果
    print(match.group())
else:
    print("no match!")
"""
print(match)
    
print(type(match))
print(rawStr)
```

```python
rawStr2 = str(bsObj.h1)


p = re.compile(r"\d{2,}") #用<h1> 的話，不可以用 \d+，因為h1的1會一起被抓出來

match = p.findall(rawStr2)
"""
if match:
    # 得到匹配結果
    print(match.group())
else:
    print("no match!")
"""
print(match)
    
print(type(match))
print(rawStr)
```
