### 不要用輕易用普通的 "=" ，這個只是參照，要用 copy()

```python
errCrawlList = copy.copy(errList)
print (errCrawlList)
```

```python
while len(errCrawlList) > 0:
    target = errCrawlList[0]
        #print(each)
    tempURL = "http://k-db.com/stocks/" + target + "/5m/20170619?download=csv"
    eachFile = urllib.URLopener()
    eachPath = "./" + todayStr + "/" + target + ".csv"
    time.sleep(1)
    try:
        eachFile.retrieve(tempURL, eachPath)
    except IOError:
        #errList.append(eachS)
        print(target + " got IOError, added to errList")
        time.sleep(3)
        
    else:
        print(target + " was done")
        errCrawlList.remove(target)
    
    #sleepTime = random.randint(1,4)
    time.sleep(1)
    
```
